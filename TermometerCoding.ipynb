```python
def termometer_coding(x):
    # Convert a list of numbers to a thermometer-like coding.
    # The input list is first normalized to a range of 0 to 8,
    # then each normalized value is converted to an integer.
    # Finally, each integer is converted to a "thermometer" value
    # by repeatedly applying `(2**number)-1` function.
    # This function prints the original value, the normalized integer value,
    # the thermometer value, and its binary representation for each element.
    #
    # Args:
    #   x: A list of numbers. (floating point)
    #
    # Returns:
    #   A list of the thermometer-coded values. (uint8)

    min_val = min(x)
    max_val = max(x)

    # Normalization
    for i in range(len(x)):
        x[i] = (x[i] - min_val) * 8 / (max_val - min_val)

    # Convert from float 0-1 to integer value 0-8
    integer_vals = []
    for _ in x:
        integer_vals.append(int(round(_)))

    # Convert to temperature coding
    t = []
    for number in integer_vals:
        t_val = (2**number) - 1
        t.append(t_val)

    # Checking the result - comment it for production
    for _ in range(len(x)):
        print(f'{x[_]:>.2f}', f'{integer_vals[_]:>3}' ,f'{t[_]:>3}', f'{t[_]:08b}')
    return t
```

```python
x = [x for x in range(1000)]
t = termometer_coding(x)
```
